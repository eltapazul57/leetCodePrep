## 1)  Contains duplicate
Given an integer array nums, return true if any value appears more than once in the array, otherwise return false.

Example 1:

Input: nums = [1, 2, 3, 3]

Output: true

Example 2:

Input: nums = [1, 2, 3, 4]

Output: false

```python

def hasDuplicate(self, nums: List[int]) -> bool:
    dictionaryCheck = {}
    for i in nums:
        if i in dictionaryCheck:
            return True
        else:
            dictionaryCheck[i] = 1
    return False
```
