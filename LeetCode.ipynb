## 1)  Contains duplicate
Given an integer array nums, return true if any value appears more than once in the array, otherwise return false.

**Example 1:**

Input: nums = [1, 2, 3, 3]

Output: true

**Example 2:**

Input: nums = [1, 2, 3, 4]

Output: false

### Solution:
Here I used a dictionary in which I add the numbers of the list if it is not there, if it is there I can simply  return true since there would be a repeated value. When I finish the  loop it means that no nomber is repeated so I return false.

````python

def hasDuplicate(nums:list) -> bool:
    dictionaryCheck = {}
    for i in nums:
        if i in dictionaryCheck:
            return True
        else:
            dictionaryCheck[i] = 1
    return False
````

## 2) Valid Anagram

Given two strings s and t, return true if the two strings are anagrams of each other, otherwise return false.

An anagram is a string that contains the exact same characters as another string, but the order of the characters can be different.

**Example 1:**

Input: s = "racecar", t = "carrace"

Output: true
**Example 2:**

Input: s = "jar", t = "jam"

Output: false

### Solution:
I used 2 list for the 2 strings and I appended each character in  those lists. Then  I sorted them, if they are equal then all good, return True, if not then return False. 

````python
def isAnagram(s: str, t: str) -> bool:
    sList = []
    tList = []
    for i in s:
        sList.append(i)
    for j in t:
        tList.append(j)
    sList.sort()
    tList.sort()
    if sList == tList:
        return True
    else:
        return False
        
````

## 3) Two Sum
Given an array of integers nums and an integer target, return the indices i and j such that nums[i] + nums[j] == target and i != j.

You may assume that every input has exactly one pair of indices i and j that satisfy the condition.

Return the answer with the smaller index first.

**Example 1:**

Input: 
nums = [3,4,5,6], target = 7

Output: [0,1]

Explanation: nums[0] + nums[1] == 7, so we return [0, 1].

**Example 2:**

Input: nums = [4,5,6], target = 10

Output: [0,2]

**Example 3:**

Input: nums = [5,5], target = 10

Output: [0,1]
**Constraints:**
- 2 <= nums.length <= 1000
- 10,000,000 <= nums[i] <= 10,000,000
- 10,000,000 <= target <= 10,000,000
Only one valid answer exists.

### Solution:
Here it is O(n²) since I have to check each  number against each number,  if the  sum is the target, all good, if not continue iterating



````python
def twoSum(nums: list, target: int) -> list:
    answer = []
    for i in range(0,len(nums)):
        for j in range(0,len(nums)):
            check = nums[i] + nums[j]
            
            if check == target and i != j:
                answer.append(i)
                answer.append(j)
                answer.sort()
                return answer
            else:
                pass



````

## 4) Group Anagrams

Given an array of strings strs, group all anagrams together into sublists. You may return the output in any order.

An anagram is a string that contains the exact same characters as another string, but the order of the characters can be different.

**Example 1:**

Input: strs = ["act","pots","tops","cat","stop","hat"]

Output: [["hat"],["act", "cat"],["stop", "pots", "tops"]]

**Example 2:**

Input: strs = ["x"]

Output: [["x"]]

**Example 3:**

Input: strs = [""]

Output: [[""]]

**Constraints:**

1 <= strs.length <= 1000.
0 <= strs[i].length <= 100
strs[i] is made up of lowercase English letters.

### Solution:
The magic is in a dictionary where the key is the letters of the anagram sorted, the value of those keys is a list of the strings that have that anagram. The key needs to be a tuple so I have to convert the list to a tuple because a touple is an inmutable object.



````python
def groupAnagrams(strs: list) -> list:
    possibleAnagrams = {}
    answer = []
    for string in strs:
        anagramCheck = []
        for character in string:
            anagramCheck.append(character)
        anagramCheck.sort()
        keyTuple = tuple(anagramCheck)
        if keyTuple not in possibleAnagrams:
            possibleAnagrams[keyTuple] = []
        possibleAnagrams[keyTuple].append(string)

    answer = list(possibleAnagrams.values())
    return answer
````

**So Machine 1 does:**

**and Machine 2 does:**

**strs2 in Machine 2 should be the same as strs in Machine 1.**

Implement the encode and decode methods.

**Example 1:**

Input: dummy_input = ["Hello","World"]

Output: ["Hello","World"]

**Explanation:**

Machine 1:

Codec encoder = new Codec();

String msg = encoder.encode(strs);

Machine 1 ---msg---> Machine 2

Machine 2:

Codec decoder = new Codec();

String[] strs = decoder.decode(msg);

**Example 2:**

Input: dummy_input = [""]

Output: [""]

**Constraints:**
- 0 <= strs.length < 100
- 0 <= strs[i].length < 200
- strs[i] contains any possible characters out of 256 valid ASCII characters.

### Solution:
First I create an empty string and I add to that string itself and a '{' as a special  character to know where each  word starts and finishes. When I decode it I check if the string is empty, if it is I return an empty list. If not i need to separate each string by the special character stated before, I need to take the string from the character on indent 1 since the one on indent 0 is a '{'

````python
def encode(self, strs: list[str]) -> str:
    string = ""
    for s in strs:
        string = string +"{" + s
    return string
        

def decode( s: str) -> list[str]:
    if s == "":
        return []
    else:    
        newString = s[1:]
        newString = newString.split("{")
        return newString
````

````python
def productExceptSelf(nums: list[int]) -> list[int]:
        answer = []
        for i in range(0,len(nums)):
            product = 1
            for j in range(0,len(nums)):
                if i == j:
                    pass
                else:
                    product = product*nums[j]
            answer.append(product)
        return answer
````

## 8) Valid Sudoku

You are given a 9 x 9 Sudoku board board. A Sudoku board is valid if the following rules are followed:

- Each row must contain the digits 1-9 without duplicates.
- Each column must contain the digits 1-9 without duplicates.
- Each of the nine 3 x 3 sub-boxes of the grid must contain the digits 1-9 without duplicates.
- Return true if the Sudoku board is valid, otherwise return false

**Note:** A board does not need to be full or be solvable to be valid.

**Example 1:**


**Input:**

|---|---|---|---|---|---|---|---|---|---|
| **1** | 1 | 2 | . | . | 3 | . | . | . | . |

| **8** | . | . | . | 4 | 1 | 9 | . | . | 8 |
| **9** | . | . | . | . | 8 | . | . | 7 | 9 |

Output: true

**Example 2:**

| | 1 | 2 | 3 | 4 | 5 | 6 | 7 | 8 | 9 |
|---|---|---|---|---|---|---|---|---|---|
| **1** | 1 | 2 | . | . | 3 | . | . | . | . |

| **6** | 7 | . | . | . | 2 | . | . | . | 6 |
| **7** | . | . | . | . | . | . | 2 | . | . |
| **8** | . | . | . | 4 | 1 | 9 | . | . | 8 |
| **9** | . | . | . | . | 8 | . | . | 7 | 9 |



Output: false

Explanation: There are two 1's in the top-left 3x3 sub-box.

**Constraints:**

board.length == 9
board[i].length == 9
board[i][j] is a digit 1-9 or '.'.

### Solution:
First  I check the columns and rows so  that there no duplicates. Then I go through each square, in each square I check for the columns and rows of it, if there are any duplicates it returns false imediately, otherwise when the loops are  completed it means that the sudoku  is valid.


````python
def isValidSudoku(board: list[list[str]])-> bool:
    for i in range(0,len(board)):
        nCheckHorizontal = {}
        nCheckVertical  = {}
        for j in range(0,len(board)):
            
            if board[i][j] not in nCheckHorizontal and  board[i][j] != '.':
                nCheckHorizontal[board[i][j]] = 1
            elif board[i][j] == '.':
                pass
            else:
                return False
            if board[j][i] not in nCheckVertical and  board[j][i] != '.':
                nCheckVertical[board[j][i]] = 1
            elif board[j][i] == '.':
                pass
            else:
                return False
            
        for row in range(0,9,3):
            for  column in range(0,9,3):
                nCheckSquare = {}
                for squareRow in range(0,3):
                    for squareColumn in range(0,3):
                        
                        if board[squareRow+row][squareColumn+column] not in nCheckSquare and board[squareRow+row][squareColumn+column] != '.':
                            nCheckSquare[board[squareRow+row][squareColumn+column]] = 1
                        elif board[squareRow+row][squareColumn+column] == '.':
                            pass
                        elif board[squareRow+row][squareColumn+column] in nCheckSquare:
                            return False
   
        return True
        
````
