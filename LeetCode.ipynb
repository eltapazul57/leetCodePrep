## 1)  Contains duplicate
Given an integer array nums, return true if any value appears more than once in the array, otherwise return false.

**Example 1:**

Input: nums = [1, 2, 3, 3]

Output: true

**Example 2:**

Input: nums = [1, 2, 3, 4]

Output: false

### Solution:
Here I used a dictionary in which I add the numbers of the list if it is not there, if it is there I can simply  return true since there would be a repeated value. When I finish the  loop it means that no nomber is repeated so I return false.

```python

def hasDuplicate(nums:list) -> bool:
    dictionaryCheck = {}
    for i in nums:
        if i in dictionaryCheck:
            return True
        else:
            dictionaryCheck[i] = 1
    return False
```

## 2) Valid Anagram

Given two strings s and t, return true if the two strings are anagrams of each other, otherwise return false.

An anagram is a string that contains the exact same characters as another string, but the order of the characters can be different.

**Example 1:**

Input: s = "racecar", t = "carrace"

Output: true
**Example 2:**

Input: s = "jar", t = "jam"

Output: false

### Solution:
I used 2 list for the 2 strings and I appended each character in  those lists. Then  I sorted them, if they are equal then all good, return True, if not then return False. 

```python
def isAnagram(s: str, t: str) -> bool:
    sList = []
    tList = []
    for i in s:
        sList.append(i)
    for j in t:
        tList.append(j)
    sList.sort()
    tList.sort()
    if sList == tList:
        return True
    else:
        return False
        
```

## 3) Two Sum
Given an array of integers nums and an integer target, return the indices i and j such that nums[i] + nums[j] == target and i != j.

You may assume that every input has exactly one pair of indices i and j that satisfy the condition.

Return the answer with the smaller index first.

**Example 1:**

Input: 
nums = [3,4,5,6], target = 7

Output: [0,1]

Explanation: nums[0] + nums[1] == 7, so we return [0, 1].

**Example 2:**

Input: nums = [4,5,6], target = 10

Output: [0,2]

**Example 3:**

Input: nums = [5,5], target = 10

Output: [0,1]
**Constraints:**
- 2 <= nums.length <= 1000
- 10,000,000 <= nums[i] <= 10,000,000
- 10,000,000 <= target <= 10,000,000
Only one valid answer exists.

### Solution:
Here it is O(n²) since I have to check each  number against each number,  if the  sum is the target, all good, if not continue iterating



```python
def twoSum(nums: list, target: int) -> list:
    answer = []
    for i in range(0,len(nums)):
        for j in range(0,len(nums)):
            check = nums[i] + nums[j]
            
            if check == target and i != j:
                answer.append(i)
                answer.append(j)
                answer.sort()
                return answer
            else:
                pass



```

## 4) Group Anagrams

Given an array of strings strs, group all anagrams together into sublists. You may return the output in any order.

An anagram is a string that contains the exact same characters as another string, but the order of the characters can be different.

**Example 1:**

Input: strs = ["act","pots","tops","cat","stop","hat"]

Output: [["hat"],["act", "cat"],["stop", "pots", "tops"]]

**Example 2:**

Input: strs = ["x"]

Output: [["x"]]

**Example 3:**

Input: strs = [""]

Output: [[""]]

**Constraints:**

1 <= strs.length <= 1000.
0 <= strs[i].length <= 100
strs[i] is made up of lowercase English letters.

### Solution:
I was not able to solve it by myself. The reasoning for the solution is: Create a default dict, why a default dict, so that I do not have to deal if the key is not created. Then I iterate through each string of the given strings and I create a count of the used letters, a list of 0 to 25 (26 zeroes) each index representing a letter of the alphabet. Then we need to iterate through each character of each string and count how many times each character appears. Thats when ASCII (ord) comes in handy, it allows to identify the index of the letters, it is needed to delete the ord(a) so that the letters are in a 0-25 range, there is a +=1 because if there is a repeated letter we want to add 1 to the count. Then we use a tuple because it is an inmutable object in python. When we have the touple (1,1,1,0...0) we add all the strings to that.



```python
from collections import defaultdict


def groupAnagrams2(strs: list) -> list:
    possibleAnagrams = defaultdict(list)
    for string in strs:
        count = [0]*26
        for character in string:
            count[ord(character)-ord("a")]+=1
        possibleAnagrams[tuple(count)].append(string)
    return possibleAnagrams.values()
```

```python
def groupAnagrams(strs: list) -> list:
    possibleAnagrams = {}
    answer = []
    for string in strs:
        anagramCheck = []
        for character in string:
            anagramCheck.append(character)
        anagramCheck.sort()
        keyTuple = tuple(anagramCheck)    
        possibleAnagrams[keyTuple].append(string)
    
    copyStrs  = strs.copy()
    
    for anagram in possibleAnagrams:
        tempAnswer = []
        if anagram not in strs:
            pass
        else:
            for string in copyStrs:
                if possibleAnagrams[anagram] == possibleAnagrams[string] and anagram != string:
                    tempAnswer.append(string)
                    strs.remove(string)
            strs.remove(anagram)
            tempAnswer.append(anagram)
            answer.append(tempAnswer)
            
    return answer
```
