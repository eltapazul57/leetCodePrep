## 1)  Contains duplicate
Given an integer array nums, return true if any value appears more than once in the array, otherwise return false.

**Example 1:**

Input: nums = [1, 2, 3, 3]

Output: true

**Example 2:**

Input: nums = [1, 2, 3, 4]

Output: false

### Solution:
Here I used a dictionary in which I add the numbers of the list if it is not there, if it is there I can simply  return true since there would be a repeated value. When I finish the  loop it means that no nomber is repeated so I return false.

````python

def hasDuplicate(nums:list) -> bool:
    dictionaryCheck = {}
    for i in nums:
        if i in dictionaryCheck:
            return True
        else:
            dictionaryCheck[i] = 1
    return False
````

## 2) Valid Anagram

Given two strings s and t, return true if the two strings are anagrams of each other, otherwise return false.

An anagram is a string that contains the exact same characters as another string, but the order of the characters can be different.

**Example 1:**

Input: s = "racecar", t = "carrace"

Output: true
**Example 2:**

Input: s = "jar", t = "jam"

Output: false

### Solution:
I used 2 list for the 2 strings and I appended each character in  those lists. Then  I sorted them, if they are equal then all good, return True, if not then return False. 

````python
def isAnagram(s: str, t: str) -> bool:
    sList = []
    tList = []
    for i in s:
        sList.append(i)
    for j in t:
        tList.append(j)
    sList.sort()
    tList.sort()
    if sList == tList:
        return True
    else:
        return False
        
````

## 3) Two Sum
Given an array of integers nums and an integer target, return the indices i and j such that nums[i] + nums[j] == target and i != j.

You may assume that every input has exactly one pair of indices i and j that satisfy the condition.

Return the answer with the smaller index first.

**Example 1:**

Input: 
nums = [3,4,5,6], target = 7

Output: [0,1]

Explanation: nums[0] + nums[1] == 7, so we return [0, 1].

**Example 2:**

Input: nums = [4,5,6], target = 10

Output: [0,2]

**Example 3:**

Input: nums = [5,5], target = 10

Output: [0,1]
**Constraints:**
- 2 <= nums.length <= 1000
- 10,000,000 <= nums[i] <= 10,000,000
- 10,000,000 <= target <= 10,000,000
Only one valid answer exists.

### Solution:
Here it is O(n²) since I have to check each  number against each number,  if the  sum is the target, all good, if not continue iterating



````python
def twoSum(nums: list, target: int) -> list:
    answer = []
    for i in range(0,len(nums)):
        for j in range(0,len(nums)):
            check = nums[i] + nums[j]
            
            if check == target and i != j:
                answer.append(i)
                answer.append(j)
                answer.sort()
                return answer
            else:
                pass



````

## 4) Group Anagrams

Given an array of strings strs, group all anagrams together into sublists. You may return the output in any order.

An anagram is a string that contains the exact same characters as another string, but the order of the characters can be different.

**Example 1:**

Input: strs = ["act","pots","tops","cat","stop","hat"]

Output: [["hat"],["act", "cat"],["stop", "pots", "tops"]]

**Example 2:**

Input: strs = ["x"]

Output: [["x"]]

**Example 3:**

Input: strs = [""]

Output: [[""]]

**Constraints:**

1 <= strs.length <= 1000.
0 <= strs[i].length <= 100
strs[i] is made up of lowercase English letters.

### Solution:
The magic is in a dictionary where the key is the letters of the anagram sorted, the value of those keys is a list of the strings that have that anagram. The key needs to be a tuple so I have to convert the list to a tuple because a touple is an inmutable object.



````python
def groupAnagrams(strs: list) -> list:
    possibleAnagrams = {}
    answer = []
    for string in strs:
        anagramCheck = []
        for character in string:
            anagramCheck.append(character)
        anagramCheck.sort()
        keyTuple = tuple(anagramCheck)
        if keyTuple not in possibleAnagrams:
            possibleAnagrams[keyTuple] = []
        possibleAnagrams[keyTuple].append(string)

    answer = list(possibleAnagrams.values())
    return answer
````

## 5) Top K frecuent elements

Given an integer array nums and an integer k, return the k most frequent elements within the array.

The test cases are generated such that the answer is always unique.

You may return the output in any order.

**Example 1:**

Input: nums = [1,2,2,3,3,3], k = 2

Output: [2,3]

**Example 2:**

Input: nums = [7,7], k = 1

Output: [7]

**Constraints:**

- 1 <= nums.length <= 10^4.
- -1000 <= nums[i] <= 1000
- 1 <= k <= number of distinct elements in nums.

### Solution: 
Here I have a dictionary with number:frecuency so that i can easily count the number of times a number appears, then I sort it by descending order and then I go in a for loop from 0 to K getting those values.


````python
def topKFrequent(nums: list, k: int) -> list:
    numberCount = {} 
    answer = []
    for number in nums:
        if number not in numberCount:
            numberCount[number] = 0
        numberCount[number] += 1
    print("number count: ")
    print(numberCount)
    numberCountSorted = dict(sorted(numberCount.items(), key=lambda item:item[1], reverse=True))
    print("Number Count sorted: ")
    print(numberCountSorted)

    for item in list(numberCountSorted)[0:k]:
        answer.append(numberCountSorted[item])
    return answer
````
